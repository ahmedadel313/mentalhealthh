```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
upload = files.upload()
```

```python
data = pd.read_csv("Mental Health Awareness in College Students (Responses) - Form Responses 1.csv")
Academic_Stress_Frequancy = data["Please indicate how frequently you experience stress related to academic pressures"]
reason_counts = Academic_Stress_Frequancy.value_counts()
pie_chart_slice_labels = reason_counts.index.to_numpy()
pie_chart_slice_values = reason_counts.to_numpy()
plt.figure(figsize=(8, 8))
plt.pie(pie_chart_slice_values, labels=pie_chart_slice_labels, autopct="%1.1f%%", startangle=140)
plt.title("Academic Stress Frequancy")
plt.tight_layout()
plt.show()
```

```python
mental_health_support_utilization = data["Have you utilized mental health support services provided by your college campus?"]
reason_counts = mental_health_support_utilization.value_counts()
bar_chart_labels = reason_counts.index.to_numpy()
bar_chart_values = reason_counts.to_numpy()
plt.figure(figsize=(6, 6))
plt.bar(bar_chart_labels, bar_chart_values, color='skyblue')
plt.xlabel('Utilized Mental Health Support Services')
plt.ylabel('Frequency')
plt.title('Mental Health Support Utilization')
plt.tight_layout()
plt.show()
```

```python
discussion_feelings = data["How open do you feel discussing mental health concerns with friends or family?"]
feeling_counts = discussion_feelings.value_counts()
response_order = [
    "Not at all comfortable",
    "Somewhat uncomfortable",
    "Neutral",
    "Somewhat comfortable",
    "Very comfortable"
]
feeling_counts = feeling_counts.reindex(response_order)
bar_chart_labels = feeling_counts.index.to_numpy()
bar_chart_values = feeling_counts.to_numpy()
plt.figure(figsize=(8, 6))
plt.barh(bar_chart_labels, bar_chart_values, color='skyblue')
plt.xlabel('Frequency')
plt.ylabel('Openness Level')
plt.title('Openness in Discussing Mental Health Concerns with Friends or Family')
plt.tight_layout()
plt.show()
```

```python
GPA = data["What is your current cumulative GPA (Grade Point Average)?"]
reason_counts = GPA.value_counts()
pie_chart_slice_labels = reason_counts.index.to_numpy()
pie_chart_slice_values = reason_counts.to_numpy()
plt.figure(figsize=(8, 8))
plt.pie(pie_chart_slice_values, labels=pie_chart_slice_labels, autopct="%1.1f%%", startangle=140)
plt.title("GPA")
plt.tight_layout()
plt.show()
```

```python
anxiety_dep_symp = data["Have you experienced symptoms of anxiety or depression since beginning college?"]
reason_counts = anxiety_dep_symp.value_counts()
bar_chart_labels = reason_counts.index.to_numpy()
bar_chart_values = reason_counts.to_numpy()
plt.figure(figsize=(6, 6))
plt.bar(bar_chart_labels, bar_chart_values, color='skyblue')
plt.xlabel('Symptoms Experience')
plt.ylabel('Frequency')
plt.title('Frequency of Anxiety or Depression Symptoms Since College Start')
plt.tight_layout()
plt.show()
```
